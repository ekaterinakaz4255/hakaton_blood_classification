```python
# Загружаем датасет с анализами CBC_L.xpt  #https://wwwn.cdc.gov/Nchs/Nhanes/2021-2022/DEMO_L.htm
file_path = "CBC_L.xpt"
df_cbc, meta = pyreadstat.read_xport(file_path)

# Вывод информации о датасете
print(f"Количество строк: {df_cbc.shape[0]}")
print(f"Количество столбцов: {df_cbc.shape[1]}")
print(f"Названия столбцов: {list(df_cbc.columns)}")
```

```python
 # SEQN - Respondent sequence number
    # WTPH2YR - Phlebotomy 2 Year Weight
    # LBXWBCSI - White blood cell count (1000 cells/uL)  +++
    # LBXLYPCT - Lymphocyte percent (%)
    # LBXMOPCT - Monocyte percent (%)
    # LBXNEPCT - Segmented neutrophils percent (%)
    # LBXEOPCT - Eosinophils percent (%)
    # LBXBAPCT - Basophils percent (%)
    # LBDLYMNO - Lymphocyte number (1000 cells/uL)
    # LBDMONO - Monocyte number (1000 cells/uL)
    # LBDNENO - Segmented neutrophils num (1000 cell/uL)
    # LBDEONO - Eosinophils number (1000 cells/uL)
    # LBDBANO - Basophils number (1000 cells/uL)
    # LBXRBCSI - Red blood cell count (million cells/uL) +++
    # LBXHGB - Hemoglobin (g/dL)
    # LBXHCT - Hematocrit (%)
    # LBXMCVSI - Mean cell volume (fL)
    # LBXMC - Mean Cell Hgb Conc. (g/dL)
    # LBXMCHSI - Mean cell hemoglobin (pg)
    # LBXRDW - Red cell distribution width (%)
    # LBXPLTSI - Platelet count (1000 cells/uL)   
    # LBXMPSI - Mean platelet volume (fL)
    # LBXNRBC - Nucleated red blood cells
```

```python
# Загружаем датасет с информацией об участниках DEMO_L.xpt
file_path = "DEMO_L.xpt"
df_demo, meta = pyreadstat.read_xport(file_path, encoding="latin1")  # Указываем кодировку

# Вывод информации о датасете
print(f"Количество строк: {df_demo.shape[0]}")
print(f"Количество столбцов: {df_demo.shape[1]}")
print(f"Названия столбцов: {list(df_demo.columns)}")
```

```python
# Удаление беременных из датасета
df_demo = df_demo[df_demo['RIDEXPRG'] != 1]
```

```python
# Находим пересечения по колонке SEQN
common_seqn = set(df_cbc['SEQN']).intersection(set(df_demo['SEQN']))

# Выводим количество пересечений
print(f"Количество пересечений по SEQN: {len(common_seqn)}")
```

- Переименовываем колонкb RIAGENDR -> Sex, RIDAGEYR -> Age
  
- Добавляем эти признаки в датасет с анализами по SEQN, переименовываем SEQN -> ID

- Заменяем в SEX 2 на 1 - женщины и 1 на 0 - мужчины

```python

# Переименовываем колонки в df_demo
df_demo.rename(columns={'RIAGENDR': 'Sex', 'RIDAGEYR': 'Age'}, inplace=True)

# Оставляем только нужные колонки из df_demo для присоединения
df_demo_filtered = df_demo[['SEQN', 'Sex', 'Age']]

# Выполняем объединение по колонке SEQN
df_cbc = df_cbc.merge(df_demo_filtered, on='SEQN', how='left')

# Заменяем значения в колонке SEX
df_cbc['Sex'] = df_cbc['Sex'].replace({2: 1, 1: 0})

# Заменяем название колонки SEQN на ID
df_cbc.rename(columns={'SEQN': 'ID'}, inplace=True)

print(f"Количество строк: {df_cbc.shape[0]}")
print(f"Количество столбцов: {df_cbc.shape[1]}")
print(f"Названия столбцов: {list(df_cbc.columns)}")
```

- LBXWBCSI - White blood cell count (1000 cells/uL) -> WBC 10^9/л (Лейкоциты)

- LBXRBCSI - Red blood cell count (million cells/uL) -> RBC 10^12/л (Эритроциты)

- LBXHGB - Hemoglobin (g/dL) *10 -> HGB g/L (Гемоглобин)
  
- LBXHCT - Hematocrit (%) -> HCT % (Гематокрит)

- LBXMCVSI - Mean cell volume (fL) - MCV фл (Средний объем эритроцита)
  
- LBXMC - Mean Cell Hgb Conc. (g/dL) *10 -> MCHC г/л (Средняя концентрация гемоглобина в эритроците)

- LBXMCHSI - Mean cell hemoglobin (pg) -> MCH пг (Среднее содержание гемоглобина в эритроците)

- LBXRDW - Red cell distribution width (%) % -> RDW % (Распределение эритроцитов по объему)

- LBXPLTSI - Platelet count (1000 cells/uL) -> PLT 10^9/л (Тромбоциты) 

- LBXMPSI - Mean platelet volume (fL) -> MPV фл (Средний объем тромбоцита) 

- LBXLYPCT - Lymphocyte percent (%) -> LY_REL % (Лимфоциты, относительное количество)

- LBXMOPCT - Monocyte percent (%) -> MO_REL (Моноциты, относительное количество)

- LBXNEPCT - Segmented neutrophils percent (%) -> NE_REL (Нейтрофилы, относительное количество)

- LBXEOPCT - Eosinophils percent (%) -> EO_REL (Эозинофилы, относительное количество)

- LBXBAPCT - Basophils percent (%) -> BA_REL (Базофилы, относительное количество)

- LBDLYMNO - Lymphocyte number (1000 cells/uL) -> LY_ABS 10^9/л (Лимфоциты, абсолютное количество) 

- LBDMONO - Monocyte number (1000 cells/uL) -> MO_ABS 10^9/л (Моноциты, абсолютное количество)

- LBDNENO - Segmented neutrophils num (1000 cell/uL) -> NE_ABS 10^9/л (Нейтрофилы, абсолютное количество) 
  
- LBDEONO - Eosinophils number (1000 cells/uL) -> EO_ABS 10^9/л (Эозинофилы, абсолютное количество)
   
- LBDBANO - Basophils number (1000 cells/uL) -> BA_ABS (Базофилы, абсолютное количество)

- Переименовываем колонки в соответствии с описанием, домножаем значения колонок LBXMC и LBXHGB на 10.

- удаляем колонки WTPH2YR и LBXNRBC

```python

# Переименовываем колонки в соответствии с описанием
df_cbc.rename(columns={
    'LBXWBCSI': 'WBC',  # Лейкоциты
    'LBXRBCSI': 'RBC',  # Эритроциты
    'LBXHGB': 'HGB',    # Гемоглобин
    'LBXHCT': 'HCT',    # Гематокрит
    'LBXMCVSI': 'MCV',  # Средний объем эритроцита
    'LBXMC': 'MCHC',    # Средняя концентрация гемоглобина в эритроците
    'LBXMCHSI': 'MCH',  # Среднее содержание гемоглобина в эритроците
    'LBXRDW': 'RDW',    # Распределение эритроцитов по объему
    'LBXPLTSI': 'PLT',  # Тромбоциты
    'LBXMPSI': 'MPV',   # Средний объем тромбоцита
    'LBXLYPCT': 'LY_REL',  # Лимфоциты, относительное количество
    'LBXMOPCT': 'MO_REL',  # Моноциты, относительное количество
    'LBXNEPCT': 'NE_REL',  # Нейтрофилы, относительное количество
    'LBXEOPCT': 'EO_REL',  # Эозинофилы, относительное количество
    'LBXBAPCT': 'BA_REL',  # Базофилы, относительное количество
    'LBDLYMNO': 'LY_ABS',  # Лимфоциты, абсолютное количество
    'LBDMONO': 'MO_ABS',   # Моноциты, абсолютное количество
    'LBDNENO': 'NE_ABS',   # Нейтрофилы, абсолютное количество
    'LBDEONO': 'EO_ABS',   # Эозинофилы, абсолютное количество
    'LBDBANO': 'BA_ABS'    # Базофилы, абсолютное количество
}, inplace=True)

# Домножаем колонки LBXMC и LBXHGB на 10
df_cbc['MCHC'] *= 10
df_cbc['HGB'] *= 10

# Удаляем колонки WTPH2YR и LBXNRBC
df_cbc.drop(columns=['WTPH2YR', 'LBXNRBC'], inplace=True)

# Вывод информации о датасете
print(f"Количество строк: {df_cbc.shape[0]}")
print(f"Количество столбцов: {df_cbc.shape[1]}")
print(f"Названия столбцов: {list(df_cbc.columns)}")
```
